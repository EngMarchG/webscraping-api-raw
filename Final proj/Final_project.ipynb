## Note that we took New York alone and saved it and proceeded to replace it with San Francisco. Hence, we will be loading the datas from the saved format (or at least NY)

```python
# importing the requests library
import requests

# location given here
#location = "Lebanon"

#Column placeholders
brand_ny = []
model_ny = []
mileage_ny = []
year_ny = []
price_ny = []
price_drop_ny = []

#Scrapping loop
for i in range(1,168):
    
    # api-endpoint
    URL = f'https://www.truecar.com/abp/api/vehicles/used/listings?city=san-francisco&collapse=true&fallback=true&include_incentives=true&include_targeted_incentives=true&new_or_used=u&page={i}&per_page=30&postal_code=94102&search_event=true&sort\[\]=best_match&sponsored=true&state=ca'


    # sending get request and saving the response as response object
    r = requests.get(url = URL)

    # extracting data in json format
    data = r.json()
    data_items = data["listings"]
    

    for ele in data_items:
        if i <167:
            brand_ny.append(ele["vehicle"]["make"])
            model_ny.append(ele["vehicle"]["model"])
            mileage_ny.append(ele["vehicle"]["mileage"])
            year_ny.append(ele["vehicle"]["year"])
            price_ny.append(ele["vehicle"]["list_price"])
            price_drop_ny.append(ele["vehicle"]["price_dropped"])
        
        elif i == 167 and len(brand_ny)<5001:
            brand_ny.append(ele["vehicle"]["make"])
            model_ny.append(ele["vehicle"]["model"])
            mileage_ny.append(ele["vehicle"]["mileage"])
            year_ny.append(ele["vehicle"]["year"])
            price_ny.append(ele["vehicle"]["list_price"])
            price_drop_ny.append(ele["vehicle"]["price_dropped"])
            
        else:
            break
            
        
df_ny = pd.DataFrame({"Brand": brand_ny, "Model": model_ny, "Mileage": mileage_ny, "Year": year_ny, "Price": price_ny, "Price Drop": price_drop_ny})


```

```python
df_ny.to_excel("Sf_car_listings.xlsx")
```
